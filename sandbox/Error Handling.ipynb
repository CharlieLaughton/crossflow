How to deal with unreliable computations

```python
from crossflow.tasks import FunctionTask
from crossflow.clients import Client
import numpy as np
import time
```

```python
def _mayfail(x):
    if np.random.random() < 0.1:
        raise RuntimeError('Error - something went wrong!')
    return x*x
```

```python
mayfail = FunctionTask(_mayfail)
mayfail.set_inputs(['x'])
mayfail.set_outputs(['s'])
```

```python
c = Client(n_workers=1)
c
```

```python
xs = list(range(1000))
squares = c.map(mayfail, xs)
max_retries = 10
n_retries = 0
failed = True
while failed and n_retries < max_retries:
    done = False
    while not done:
        time.sleep(1)
        done = np.all([s.done() for s in squares])
    
    failed = False
    for i in range(len(squares)):
        if squares[i].status == 'error':
            squares[i].retry()
            failed = True
    n_retries += 1

fails = [s for s in squares if s.status == 'error']
print(len(fails)) # should be zero
```
